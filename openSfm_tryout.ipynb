```python
import sqlite3

db_path = "/content/drive/MyDrive/sewer_reconstruction/sewer_project/database.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

cursor.execute("SELECT COUNT(*) FROM images;")
num_images = cursor.fetchone()[0]
print(f"Number of images in database: {num_images}")

cursor.execute("SELECT name FROM images LIMIT 5;")
image_names = [row[0] for row in cursor.fetchall()]
print("First 5 images:", image_names)

cursor.execute("""
SELECT images.name,
       LENGTH(keypoints.data)/16 AS num_keypoints
FROM images
JOIN keypoints ON images.image_id = keypoints.image_id
LIMIT 5;
""")
keypoints_data = cursor.fetchall()
print("First 5 images with keypoints count:")
for name, kp_count in keypoints_data:
    print(f"  {name}: {int(kp_count)} keypoints")

cursor.execute("SELECT COUNT(*) FROM matches;")
num_matches = cursor.fetchone()[0]
print(f"Number of matches in database: {num_matches}")

conn.close()
```
