## Sequential SfM

```python
OPENMVG_SFM_BIN = f"{base_dir}/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/"
CAMERA_SENSOR_WIDTH_DIRECTORY = f"{base_dir}/openMVG_workspace/openMVG/src/openMVG/exif/sensor_width_database"
```

```python
images_path = f"{base_dir}/video_4_frames_preprocessed"
output_path = f"{base_dir}/openMVG_output_video_4_v2"
matches_path = f"{output_path}/matches"
reconstruction_path = f"{output_path}/reconstruction_sequential"
```

```python
import os
for path in [output_path, matches_path, reconstruction_path]:
  os.makedirs(path, exist_ok=True)
```

```python
import subprocess
```

```python
def run_subprocess(command):
    process = subprocess.Popen(
        command,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        universal_newlines=True
    )
    for line in process.stdout:
        print(line, end='')

    process.wait()
    return process.returncode
```

```python
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_SfMInit_ImageListing
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_ComputeFeatures
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_ComputeMatches
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_GeometricFilter
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_SfM
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_ComputeSfM_DataColor
!chmod +x /content/drive/MyDrive/ai_project/openMVG_workspace/openMVG/Linux-x86_64-RELEASE/openMVG_main_ComputeStructureFromKnownPoses
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_SfMInit_ImageListing",
    "-i", images_path,
    "-o", matches_path,
    "-f", "2300",
    "-c", "3",
])
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_ComputeFeatures",
    "-i", f"{matches_path}/sfm_data.json",
    "-o", matches_path,
    "-m", "SIFT",
    "-f", "1"
])
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_ComputeMatches",
    "-i", f"{matches_path}/sfm_data.json",
    "-o", f"{matches_path}/matches.putative.bin",
    "-f", "1",
    "-n", "AUTO"
])
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_GeometricFilter",
    "-i", f"{matches_path}/sfm_data.json",
    "-m", f"{matches_path}/matches.putative.bin",
    "-g", "f",
    "-o", f"{matches_path}/matches.f.bin"
])
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_SfM",
    "--sfm_engine", "INCREMENTAL",
    "--input_file", f"{matches_path}/sfm_data.json",
    "--match_dir", matches_path,
    "--output_dir", reconstruction_path
])
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_ComputeSfM_DataColor",
    "-i", f"{reconstruction_path}/sfm_data.bin",
    "-o", f"{reconstruction_path}/colorized.ply"
])
```

```python
run_subprocess([
    f"{OPENMVG_SFM_BIN}/openMVG_main_ComputeStructureFromKnownPoses",
    "-i", f"{reconstruction_path}/sfm_data.bin",
    "-m", matches_path,
    "-o", f"{reconstruction_path}/robust.ply"
])
```
